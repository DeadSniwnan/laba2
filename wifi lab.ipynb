```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, svm
from sklearn import metrics
```

```python
data = pd.read_csv('wifi_localization.txt', sep='\t',  names = ["Room 1", "Room 2", "Room 3", "Room 4", "Room 5", "Room 6", "Room 7", "wifi N"])
```

Переносим файл исходных данных в df

```python
data
```

Проверям правильно ли данные преобразовались

```python
X = data.iloc[:,[0,1,2,3,4,5,6]]
Y = data.iloc[:,[7]]
```

Разделям исходный df на два, первый хранит замеры в комнатах, второй номера вайфай точек соответствующих сигналам первой таблицы

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, shuffle=True, stratify=Y)
```

Перемешиваем данные и разделяем их на тренировочные и тестовые

```python
log_r = linear_model.LogisticRegression(verbose=1)
log_r.fit(X_train, Y_train)
```

Тренируем модель

```python
Y_pred = log_r.predict(X_test)
print(Y_pred)
print(Y_test)
```

Тестируем и смотрим точность полученной модели 

```python
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
print(metrics.classification_report(Y_test, Y_pred))
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
```
